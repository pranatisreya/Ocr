```python
import os
import zipfile
import pandas as pd
from datasets import load_dataset

# Load dataset
dataset = load_dataset("Daniellau119/trad_chineseHKOCR", split="train")

# Randomly select 500 images
sample_size = 500
sampled_data = dataset.shuffle(seed=42).select(range(sample_size))

image_dir = "sample_images"
excel_filename = "ocr_data.xlsx"
zip_filename = "sample_trad_chineseHKOCR.zip"

os.makedirs(image_dir, exist_ok=True)

# Prepare data for Excel
ocr_data = []

for i, example in enumerate(sampled_data):
    # Save image
    img_filename = f"image_{i}.png"
    img_path = os.path.join(image_dir, img_filename)
    example['image'].save(img_path)

    # Store OCR text in list
    ocr_data.append([img_filename, "\n".join(map(str, example['ocr']))])

# Convert to DataFrame and save as Excel
df = pd.DataFrame(ocr_data, columns=["Image File", "OCR Text"])
df.to_excel(excel_filename, index=False)

# Zip everything
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, _, files in os.walk(image_dir):
        for file in files:
            file_path = os.path.join(root, file)
            zipf.write(file_path, os.path.relpath(file_path, image_dir))
    zipf.write(excel_filename)  # Include Excel file in the ZIP

print(f"Sample dataset saved as {zip_filename} with OCR data in {excel_filename}")
```

```python
from ultralytics import YOLO
import cv2
from google.colab.patches import cv2_imshow  # Use for Colab
import pytesseract

# Load pre-trained YOLO model
model = YOLO("yolov8n.pt")  # Use a smaller model for faster inference

# Detect text regions on a Chinese card
results = model("image_0.png", conf=0.4)

# Show detection results (Colab-safe)
results[0].show()

# Load image
image = cv2.imread("image_0.png")

# Ensure we access the first detection result
result = results[0]

pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# Extract bounding box coordinates and process text
for i, box in enumerate(result.boxes.xyxy):
    x1, y1, x2, y2 = map(int, box[:4])  # Convert coordinates to integers

    # Crop the detected text region
    cropped_text = image[y1:y2, x1:x2]

    # Show cropped text in Colab
    cv2_imshow(cropped_text)

    # Perform OCR on cropped text
    text = pytesseract.image_to_string(cropped_text, lang="chi_sim")
    print(f"Detected Text: {text.strip()}")

```

```python
import cv2
import matplotlib.pyplot as plt
from imgocr import ImgOcr

# Initialize OCR model
m = ImgOcr(use_gpu=False, is_efficiency_mode=True)  # Change to True if using GPU

# Perform OCR on the image
image_path = "image_0.png"
result = m.ocr(image_path)

# Load the image
image = cv2.imread(image_path)

# Draw detected text on the image
for i in result:
    text = i['text']
    box = i['box']  # Extract bounding box (4 corner points)

    # Convert box coordinates to bounding rectangle
    x_min = int(min(p[0] for p in box))
    y_min = int(min(p[1] for p in box))
    x_max = int(max(p[0] for p in box))
    y_max = int(max(p[1] for p in box))

    # Draw bounding box (Blue color for better visibility)
    cv2.rectangle(image, (x_min, y_min), (x_max, y_max), (0, 0, 255), 2)

    # Put detected text above the bounding box with smaller font
    font_scale = 0.3  # Reduce font size
    font_thickness = 1  # Reduce thickness
    cv2.putText(image, text, (x_min, y_min - 5), cv2.FONT_HERSHEY_SIMPLEX, font_scale, (255, 255, 255), font_thickness)

# Convert BGR (OpenCV format) to RGB (Matplotlib format)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the image with detections
plt.figure(figsize=(10, 6))
plt.imshow(image)
plt.axis("off")
plt.title("Detected Text Overlay")
plt.show()

# Print detected text
print("Detected Text:")
for i in result:
    print(i['text'])
```
